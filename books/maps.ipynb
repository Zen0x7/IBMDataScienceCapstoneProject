```python
import requests
import pandas as pd
import folium
from IPython.display import HTML, display

def get_json_from(url):
  try:
      # Hacer la solicitud GET
      response = requests.get(url)
      # Verificar si la solicitud fue exitosa
      if response.status_code == 200:
          # Convertir la respuesta a JSON
          return response.json()
      else:
          print(f"Error: {response.status_code} - {response.reason}")
  except Exception as e:
      print(f"Ha ocurrido un error: {e}")

launchpads = pd.json_normalize(get_json_from("https://api.spacexdata.com/v3/launchpads"))
# launches = pd.json_normalize(get_json_from("https://api.spacexdata.com/v3/launches"))
```

```python
map_html = []
for _, row in launchpads.iterrows():
  m = folium.Map(location=[row['location.latitude'], row['location.longitude']], zoom_start=9, width="600px", height="300px")
  marker = folium.Marker(location=[row['location.latitude'], row['location.longitude']], popup=row['location.name']).add_to(m)
  marker.add_child(folium.Popup(row["location.name"], show=True))
  map_html.append(m._repr_html_())

html_content = "<div style='display: grid; grid-template-columns: repeat(3, 1fr);  gap: 0px;'>"
for item in map_html:
    html_content += f"<div>{item}</div>"
html_content += "</div>"

# Mostrar la cuadr√≠cula en el notebook
display(HTML(html_content))
```
